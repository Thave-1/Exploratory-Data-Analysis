# Questions
- What are the average population of each continent over 50 years?
- What are the top 10 most populated countries?
- What are the least 10 populated countries?

```python
df.sort_values(by='2022 Population', ascending=False).head(10)
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
data = df.groupby('Continent').mean().sort_values(by='2022 Population', ascending=False)
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
d = data[['1970 Population', '1980 Population', '1990 Population',
       '2000 Population', '2010 Population', '2015 Population',
       '2020 Population', '2022 Population']]
dataviz = d.T
dataviz.index
```

```python
dataviz
```

```python
ax = plt.gca()
dataviz.plot(figsize=(10,5),ax=ax, xlabel='Year',ylabel ='Population',title='Average World Population')
xticks_labels = ['0','1970', '1980 ', '1990','2000', '2010', '2015','2020', '2022','9']
ax.set_xticklabels(xticks_labels)
legend=ax.legend(fontsize=10)
sns.set()
```

```python
ax1 =plt.gca()
yticks_labels = ['1970', '1980 ', '1990','2000', '2010', '2015','2020', '2022']
dataviz.plot(figsize=(10,7),kind ='barh',ax=ax1,width=0.9, xlabel='Population',ylabel ='Year',title='Average World Population') 
ax1.set_yticklabels(yticks_labels)
legend=ax1.legend(fontsize=10, bbox_to_anchor=(1.05,1), loc='upper left')

sns.set()
```

- Were all the countries consistent in their positions through out the years?


- What are the top 10 countries in the world with highest population in the year 2022?

```python
top_10 = df.sort_values(by='2022 Population', ascending = False).head(10)[['Rank','Country','2022 Population']]
```

```python
top_10.reset_index(drop=True, inplace=True)
```

```python
top_10 = top_10[['Country','2022 Population']]
```

```python
top_10.set_index('Country',inplace=True)
```

```python
top_10 = top_10.sort_values(by='2022 Population')
```

```python
ax1 = top_10.plot(kind='barh', xlabel='Population', title='Top 10 Most Populated Countries in 2022',color='green')
legend=ax1.legend(labels=['Population Size'], fontsize=8)
```

- What are the top 10 countries in the world with lowest population in the year 2022?

```python
lowest_10 = df.sort_values(by='2022 Population', ascending = False)[['Rank','Country','2022 Population']]
```

```python
lowest_10 = lowest_10.tail(14).head(10)
```

```python
lowest_10.reset_index(drop=True, inplace=True)
```

```python
lowest_10.set_index('Country',inplace=True)
```

```python
lowest_10 = lowest_10[['2022 Population']]
```

```python
lowest_10 = lowest_10.sort_values(by='2022 Population')
```

```python
ax2 = lowest_10.plot(kind='barh', xlabel='Population', title='Top 10 Least Populated Countries in 2022',color='red')
legend=ax2.legend(labels=['Population Size'], fontsize=8)
legend=ax2.legend(fontsize=10, bbox_to_anchor=(1.05,1), loc='upper left')
```

- What are the differences between the growth of the countries through the years for the above?


```python
new_df = df[['Country', '2022 Population','2020 Population', 
    '2015 Population', '2010 Population','2000 Population',
    '1990 Population', '1980 Population','1970 Population']]

top = new_df.sort_values(by='2022 Population', ascending = False).head(10)
top
```

```python
ax3 = dnew.plot(figsize=(20,15),xlabel='Year',ylabel='Population',title='Average World Population')
xticks_labels = ['2000','2010','2015','2020','2022']
ax3.set_xticklabels(xticks_labels)
legend=ax3.legend(fontsize=10)
sns.set()
```
